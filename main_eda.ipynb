```python

import time
from pymilvus import MilvusClient
from sentence_transformers import SentenceTransformer

# 2. Config Milvus & Collection
MILVUS_URI = "http://localhost:19530"
COLLECTION_NAME = "globaltech_nlp_project"

# 3. Load Model Embedding
print("Loading model embedding...")
model = SentenceTransformer("hiieu/halong_embedding")
print("Model loaded.")

# 4. Initialize Milvus Client & Load Collection
client = MilvusClient(uri=MILVUS_URI)
client.load_collection(COLLECTION_NAME)
print(f"Collection '{COLLECTION_NAME}' loaded.")

# 5. Search Function
def search_similar_news(query_text, top_k=10):
    print(f"\nQuery: '{query_text}'")
    t0 = time.time()
    
    # Embed query text
    query_vector = model.encode([query_text])
    
    # Search in Milvus
    search_res = client.search(
        collection_name=COLLECTION_NAME,
        data=query_vector,
        limit=top_k,
        output_fields=["title", "text", "url", "chunk_index"]
    )
    t1 = time.time()
    
    print(f"Search time: {(t1-t0)*1000:.2f} ms")
    print("-" * 80)
    
    # Display results nicely
    for hits in search_res:
        for i, hit in enumerate(hits):
            score = hit['distance'] # Cosine Similarity
            entity = hit['entity']
            
            print(f"#{i+1} [Score: {score:.4f}] - {entity.get('title', 'No Title')}")
            print(f"ðŸ”— {entity.get('url', '#')}")
            print(f"ðŸ“„ Content (Chunk {entity.get('chunk_index')}):")
            content_preview = entity.get('text', '')[:300].replace('\n', ' ')
            print(f"   \"{content_preview}...\"")
            print("-" * 50)
```
