```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader('./inputs/sciencerag.pdf')
documents = loader.load()
documents
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_documents(documents)

for chunk in chunks:
    print(chunk)
    print("----------------------------")
```

```python
from sentence_transformers import SentenceTransformer

embedding = SentenceTransformer("Alibaba-NLP/gte-large-en-v1.5", trust_remote_code=True)
```

```python
em = embedding.encode("My name is Nhan")
len(em)
```

```python
from pymilvus import connections, db, utility, FieldSchema, DataType, CollectionSchema, Collection

conn = connections.connect("default", host="localhost", port="19530")
```

```python
db.list_database()
```

```python
db.create_database("first_db")
```

```python
db.using_database("first_db")
```

```python
utility.list_collections()
```

```python
# schema
id_field = FieldSchema(
    name='id',
    dtype=DataType.INT64,
    is_primary=True,
    auto_id=True
)
source_field = FieldSchema(
    name='source',
    dtype=DataType.VARCHAR,
    max_length=255,
    is_primary=False,
)
page_field = FieldSchema(
    name='page',
    dtype=DataType.INT64
)
embedding_field = FieldSchema(
    name='embeddings',
    dtype=DataType.FLOAT_VECTOR,
    dim=1024,
)
content_field = FieldSchema(
    name='content',
    dtype=DataType.VARCHAR,
    max_length=2048,
)

schema = CollectionSchema(fields=[id_field, source_field, page_field, embedding_field, content_field])
```

```python
collection = Collection(name='science', schema=schema)
```

```python
utility.list_collections()
```

```python
# index
index_params = {
    'metric_type': 'L2',
    'index_type': 'HNSW',
    'params': {
        'M': 16,
        'efConstruction': 200
    }
}
```

```python
collection.create_index(field_name='embeddings', index_params=index_params)
```

```python
collection.load()
```

```python
sources = [chunk.metadata['source'] for chunk in chunks]
pages = [chunk.metadata['page'] for chunk in chunks]
page_contents = [chunk.page_content for chunk in chunks]
embeddings = [embedding.encode(chunk.page_content) for chunk in chunks]
```

```python
collection.insert([sources, pages, embeddings, page_contents])
```
